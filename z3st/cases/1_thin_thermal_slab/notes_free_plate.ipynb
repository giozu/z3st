# Case study: Free plate with temperature variation through the thickness only

## 1. Problem definition

We consider a flat plate of arbitrary in-plane shape and constant thickness `2h`.  
The plate is **free of external tractions** on all its surfaces.  
The temperature field depends only on the thickness coordinate `x`:

$$
T = T(x).
$$

Coordinate system:
- `x`: thickness direction (origin at mid-plane, surfaces at `x = ±h`).
- `y, z`: in-plane directions (arbitrary planform).

---

## 2. Assumptions

1. **Isotropic linear thermoelastic material**, with Young’s modulus `E`, Poisson’s ratio `ν`, thermal expansion coefficient `α`.  
2. **No body forces**.  
3. **Plane stress condition**:  
   $$
   \sigma_{xx} = \tau_{xy} = \tau_{xz} = \tau_{yz} = 0.
   $$
4. The only non-zero stresses are the in-plane components (independent of `y, z`):
   $$
   \sigma_{yy}(x) = \sigma_{zz}(x) = f(x).
   $$

---

## 3. Constitutive relations

For isotropic thermoelasticity:

$$
\varepsilon_{xx} = \frac{1}{E}\left(\sigma_{xx} - \nu(\sigma_{yy} + \sigma_{zz})\right) + \alpha T,
$$

$$
\varepsilon_{yy} = \frac{1}{E}\left(\sigma_{yy} - \nu(\sigma_{xx} + \sigma_{zz})\right) + \alpha T,
$$

$$
\varepsilon_{zz} = \frac{1}{E}\left(\sigma_{zz} - \nu(\sigma_{xx} + \sigma_{yy})\right) + \alpha T.
$$

Substituting the stress field hypothesis (`σyy = σzz = f(x)`, `σxx = 0`):

$$
\varepsilon_{yy} = \varepsilon_{zz} = \frac{1-\nu}{E} f(x) + \alpha T(x),
$$

$$
\varepsilon_{xx} = -\frac{2\nu}{E} f(x) + \alpha T(x).
$$

---

## 4. Compatibility conditions

With displacements depending only on `x` and all shear strains zero, the Saint-Venant compatibility equations reduce to:

$$
\frac{d^2 \varepsilon_{xx}}{dx^2} =
\frac{d^2 \varepsilon_{yy}}{dx^2} =
\frac{d^2 \varepsilon_{zz}}{dx^2} = 0.
$$

Using $\varepsilon_{yy}$ (same as $\varepsilon_{zz}$):

$$
\frac{d^2}{dx^2}\left(\frac{1-\nu}{E} f(x) + \alpha T(x)\right) = 0.
$$

Multiply by $\tfrac{E}{1-\nu}$:

$$
\frac{d^2}{dx^2}\left(f(x) + \frac{\alpha E}{1-\nu}\, T(x)\right) = 0.
$$

---

## 5. General solution for the stress field

Integrating twice:

$$
f(x) = -\frac{\alpha E}{1-\nu} T(x) + C_1 x + C_2,
$$

where `C1`, `C2` are integration constants.

Thus, the non-zero stresses are:

$$
\sigma_{yy}(x) = \sigma_{zz}(x) = f(x).
$$

---

## 6. Boundary conditions

The plate is traction-free on its **perimeter edges** (in the `y, z` plane).  
This requires that both the **resultant force** and **bending moment** vanish:

$$
N = \int_{-h}^{h} f(x)\, dx = 0, 
\qquad
M = \int_{-h}^{h} x f(x)\, dx = 0.
$$

These two equations determine `C2` and `C1`, respectively.

### 6.1 Derivation of C1 and C2

Let
$$
f(x) = -A\,T(x) + C_1 x + C_2, \qquad A \equiv \frac{\alpha E}{1-\nu}.
$$

Define the integrals
$$
I_0 \equiv \int_{-h}^{h} T(\zeta)\,d\zeta, 
\qquad
I_1 \equiv \int_{-h}^{h} \zeta\,T(\zeta)\,d\zeta.
$$

**Resultant $N=0$.** Using $\int_{-h}^{h} x\,dx = 0$ and $\int_{-h}^{h} dx = 2h$:
$$
0 = \int_{-h}^{h} f(x)\,dx 
= \int_{-h}^{h}\!\big[-A\,T(x) + C_1 x + C_2\big]dx
= -A I_0 + C_1\!\underbrace{\int_{-h}^{h} x\,dx}_{0} + C_2 (2h).
$$
Hence
$$
C_2 = \frac{A}{2h}\, I_0.
$$

**Bending moment $M=0$.** Using $\int_{-h}^{h} x\,dx = 0$ and $\int_{-h}^{h} x^2\,dx = \tfrac{2}{3}h^3$:
$$
0 = \int_{-h}^{h} x\,f(x)\,dx
= \int_{-h}^{h}\!\big[-A\,x\,T(x) + C_1 x^2 + C_2 x\big]dx
= -A I_1 + C_1\!\int_{-h}^{h} x^2\,dx + C_2\!\underbrace{\int_{-h}^{h} x\,dx}_{0}.
$$
Thus
$$
C_1 = A\,\frac{3}{2h^{3}}\, I_1.
$$

Substituting $C_1, C_2$ back into $f(x)$ gives
$$
f(x)=\frac{\alpha E}{1-\nu}\left[
-\,T(x) + \frac{1}{2h}\int_{-h}^{h} T(\zeta)\,d\zeta
+ \frac{3x}{2h^{3}}\int_{-h}^{h} \zeta\,T(\zeta)\,d\zeta
\right],
$$
so that $\sigma_{yy}(x)=\sigma_{zz}(x)=f(x)$.

*Notes.*  
- If $T(x)$ is even, then $I_1=0$ and $C_1=0$ (no antisymmetric part).  
- If $T(x)$ is linear, $T(x)=a+bx$, then $I_1 = b\int x^2 dx = \tfrac{2}{3} b h^3$ and the bracket cancels identically, yielding zero stress everywhere.


---

## 7. Final expression

After enforcing the boundary conditions, the stress distribution is:

$$
\sigma_{yy}(x) = \sigma_{zz}(x) =
\frac{\alpha E}{1-\nu}\left[
-\,T(x) + \frac{1}{2h}\int_{-h}^{h} T(\zeta)\,d\zeta
+ \frac{3x}{2h^3}\int_{-h}^{h} \zeta T(\zeta)\,d\zeta
\right].
$$

Note: the surfaces at `x = ±h` are automatically traction-free since `σxx = τx* = 0`.

---

## 8. Special cases

- **Linear temperature profile**:

$$
T(x) = a + b x \quad \Rightarrow \quad 
\sigma_{yy}(x) = \sigma_{zz}(x) = 0.
$$

No stresses are induced if the temperature varies linearly.

- **Quadratic temperature profile**:

$$
T(x) = T_0 + \beta x + \gamma x^2 \quad \Rightarrow \quad
\sigma_{yy}(x) = \sigma_{zz}(x) =
\frac{\alpha E}{1-\nu}\, \gamma \left(\frac{h^2}{3} - x^2\right).
$$

Stress is compressive in the hotter regions, tensile in cooler ones.

- **Cubic temperature profile**:

$$
T(x) = a + b x + c x^2 + d x^3 \quad \Rightarrow \quad
\sigma_{yy}(x) = \sigma_{zz}(x) =
\frac{\alpha E}{1-\nu}\left[c\left(\tfrac{h^2}{3} - x^2\right) 
+ d\left(\tfrac{3x}{2h^3}\int_{-h}^h \zeta^2\,d\zeta - x^3\right)\right].
$$

Here the odd term (`d x^3`) generates an antisymmetric stress component.

---

## 9. Physical interpretation

- Only **nonlinear temperature distributions** across the thickness induce stresses.  
- A linear variation leads to uniform expansion and free thermal curvature, stress-free.  
- Quadratic or higher-order variations cause **self-equilibrated stresses** balancing internally.

---


```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# material constants
k = 18.5  # (W/m-K)
Ti = 490.0  # (K)
To = 480.0  # (K)

nu = 0.28  # Poisson's ratio
E = 1.8e11  # Young's modulus (Pa)
alpha = 1.4e-5  # linear thermal expansion coefficient (1/K)

q0 = 9e4  # (W/m³)
mu = 59.0  # (1/m)

# half-thickness
h = 0.010  # (m)


# temperature distribution T(x)
def T(x):
    beta, gamma = 1e4, 1e-5
    # return To + beta*x + gamma*x**2
    Lx = 2 * h
    x_loc = x + h
    # return To + q0 / (k * mu**2) * (1 - np.exp(-mu * x_loc) - mu * x_loc * np.exp(-mu * Lx))
    return To - 3 / 0.02 * x_loc


# integrals
I0, _ = quad(T, -h, h)  # ∫ T(z) dz
I1, _ = quad(lambda z: z * T(z), -h, h)  # ∫ z T(z) dz


def sigma(x):
    """Return sigma_yy = sigma_zz at coordinate x."""
    prefactor = alpha * E / (1.0 - nu)
    return prefactor * (-T(x) + (1.0 / (2 * h)) * I0 + (3 * x / (2 * h**3)) * I1)


# sample points across thickness
xs = np.linspace(-h, h, 200)
sigmas = np.array([sigma(x) for x in xs])
temps = np.array([T(x) for x in xs])

# primary axis for stresses
fig, ax1 = plt.subplots(figsize=(7, 4))

color1 = "blue"
ax1.set_xlabel("x (mm)")
ax1.set_ylabel("Stress (MPa)", color=color1)
ax1.plot(xs * 1e3, sigmas / 1e6, color=color1, label="σyy (x) = σzz (x)")
ax1.tick_params(axis="y", labelcolor=color1)
ax1.axhline(0, color="k", linewidth=0.8, linestyle="--")
# ax1.set_ylim(-30, 30)

# secondary axis for temperature
ax2 = ax1.twinx()
color2 = "red"
ax2.set_ylabel("Temperature (K)", color=color2)
ax2.plot(xs * 1e3, temps, color=color2, linestyle=":", label="T(x)")
ax2.tick_params(axis="y", labelcolor=color2)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper center")

plt.title("Thermal stress and temperature across thickness")
fig.tight_layout()
plt.show()
```
