## Cylindrical tubes under internal and external pressure

The mechanical stresses according to **Lamé** for an infinitely long cylinder closed at the ends are

$$
\sigma_r = A - \dfrac{B}{r^2}
$$

$$
\sigma_\vartheta = A + \dfrac{B}{r^2}
$$

$$
\sigma_z = \frac{\sigma_r + \sigma_\vartheta}{2} = A
$$

where

$$
A = \dfrac{p_i R_i^2 - p_o R_o^2}{R_o^2-R_i^2}
$$

$$
B = (p_i - p_o) \dfrac{R_i^2 R_o^2}{R_o^2-R_i^2} 
$$

### Note on axial stress

In a **closed-end cylinder**, the axial stress  is generated by the pressure acting on the internal and external end caps. This stress must satisfy the global equilibrium of forces along the longitudinal axis:

$$
\sigma_z \cdot \pi (R_o^2 - R_i^2) = p_i \cdot \pi R_i^2 - p_o \cdot \pi R_o^2
$$

Solving for , we obtain:

$$
\sigma_z = \dfrac{p_i R_i^2 - p_o R_o^2}{R_o^2 - R_i^2}
$$

Analytically, this confirms that  is independent of the radius  and is exactly equal to the constant term  in the Lamé equations. Furthermore, it satisfies the condition , which is consistent with the **generalized plane strain** assumption where the axial deformation remains uniform across the entire cross-section.

It is convenient to expand and write:

$$
A = p_i \dfrac{R_i^2}{R_o^2-R_i^2} - p_o \dfrac{R_o^2}{R_o^2-R_i^2} = A_i + A_o
$$

$$
B = p_i \dfrac{R_i^2 R_o^2}{R_o^2-R_i^2} - p_o \dfrac{R_i^2 R_o^2}{R_o^2-R_i^2} = B_i + B_o
$$


The linear averages for $\sigma_\vartheta$ is:

$$
\langle \sigma_\vartheta \rangle_{\text{lin}} = \frac{1}{R_o - R_i} \int_{R_i}^{R_o} \sigma_\vartheta(r) \, dr = A + \dfrac{B}{R_o R_i}
$$

$$
\langle \sigma_\vartheta \rangle_{\text{lin}} = \dfrac{p_i R_i^2 - p_o R_o^2}{R_o^2-R_i^2} + (p_i - p_o) \dfrac{R_i^2 R_o^2}{R_o^2-R_i^2} \dfrac{1}{R_o R_i}
$$

$$
\langle \sigma_\vartheta \rangle_{\text{lin}} = \dfrac{p_i R_i^2 - p_o R_o^2 + (p_i - p_o) R_i R_o}{(R_o - R_i)(R_o + R_i)}
$$

$$
\langle \sigma_\vartheta \rangle_{\text{lin}} = \dfrac{p_i R_i (R_i + Ro) - p_o R_o (R_o + R_i)}{(R_o - R_i)(R_o + R_i)}
$$

$$
\langle \sigma_\vartheta \rangle_{\text{lin}} = \dfrac{p_i R_i - p_o R_o}{R_o - R_i} = \dfrac{p_i R_i}{t} - \dfrac{p_o R_o}{t}
$$

while for $\sigma_r$ is:

$$
\langle \sigma_r \rangle_{\text{lin}} = \frac{1}{R_o - R_i} \int_{R_i}^{R_o} \sigma_r(r) \, dr = A - \dfrac{B}{R_o R_i}
$$

$$
\langle \sigma_r \rangle_{\text{lin}} = \dfrac{p_i R_i^2 - p_o R_o^2}{R_o^2-R_i^2} - (p_i - p_o) \dfrac{R_i^2 R_o^2}{R_o^2-R_i^2} \dfrac{1}{R_o R_i}
$$

$$
\langle \sigma_r \rangle_{\text{lin}} = \dfrac{p_i R_i (R_i - Ro) - p_o R_o (R_o - R_i)}{(R_o - R_i)(R_o + R_i)}
$$

$$
\langle \sigma_r \rangle_{\text{lin}} = \dfrac{- p_i R_i - p_o R_o}{(R_o + R_i)}
$$

$$
\langle \sigma_r \rangle_{\text{lin}} = -\dfrac{p_i R_i}{(R_o + R_i)} - \dfrac{p_o R_o}{(R_o + R_i)}
$$

Lastly, for $\sigma_z$ is:
$$
\langle \sigma_z \rangle_{\text{lin}} = \frac{1}{R_o - R_i} \int_{R_i}^{R_o} A \, dr = \frac{A(R_o - R_i)}{R_o - R_i} = A
$$

That is the average of a constant. As a result:
$$
\langle \sigma_z \rangle_{\text{lin}} = \frac{p_i R_i^2 - p_o R_o^2}{R_o^2 - R_i^2}
$$



The cylindrical averages for $\sigma_\vartheta$ and $\sigma_r$ are:

$$
\langle \sigma \rangle_{\text{cyl}} = \frac{1}{\pi(R_o^2 - R_i^2)} \int_{R_i}^{R_o} \sigma(r) \cdot 2\pi r \, dr = \frac{2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} \sigma(r) \, r \, dr
$$

$$
\langle \sigma \rangle_{\text{cyl}} = A + \dfrac{2 B}{R_o^2 - R_i^2} \ln\left(\dfrac{R_o}{R_i}\right)
$$

$$
\langle \sigma_\vartheta \rangle_{\text{cyl}} = \frac{p_i R_i^2 - p_o R_o^2}{R_o^2 - R_i^2} + \frac{2 (p_i - p_o) R_i^2 R_o^2}{(R_o^2 - R_i^2)^2} \ln\left(\frac{R_o}{R_i}\right)
$$


$$
\langle \sigma_r \rangle_{\text{cyl}} = \frac{p_i R_i^2 - p_o R_o^2}{R_o^2 - R_i^2} - \frac{2 (p_i - p_o) R_i^2 R_o^2}{(R_o^2 - R_i^2)^2} \ln\left(\frac{R_o}{R_i}\right)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# ----------
# Parameters
# ----------
E = 210e9       # Young's modulus (Pa)
nu = 0.3        # Poisson's ratio
alpha = 1e-5    # Linear thermal expansion coefficient (1/K)

Ri = 1.00       # inner radius (m)
Ro = 1.50       # outer radius (m)

t = Ro-Ri

pi = 0       # internal pressure (Pa)
po = 5e6       # external pressure (Pa)

# radial coordinate
rvals = np.linspace(Ri, Ro, 150)
```

```python
# -------------
# Lamé formulas
# -------------
A = (pi * Ri**2 - po * Ro**2) / (Ro**2 - Ri**2)
B = Ri**2 * Ro**2 * (pi - po) / (Ro**2 - Ri**2)

def mechanical_stresses(r):
    # Lamè
    sigma_r = A - B / r**2
    sigma_t = A + B / r**2
    sigma_z = A * np.ones_like(r)
    return sigma_r, sigma_t, sigma_z

sigma_r_mech, sigma_t_mech, sigma_z_mech = mechanical_stresses(rvals)

def mechanical_stresses_linear_average(r):
    # sigma_r = (-pi*Ri-po*Ro)/(Ro+Ri) * np.ones_like(r)
    sigma_r = A - B / (Ri*Ro) * np.ones_like(r)
    # sigma_t = (pi*Ri-po*Ro)/t * np.ones_like(r)
    sigma_t = A + B / (Ri*Ro) * np.ones_like(r)
    sigma_z = A * np.ones_like(r) # GPS
    return sigma_r, sigma_t, sigma_z

sigma_r_mech_lin, sigma_t_mech_lin, sigma_z_mech_lin = mechanical_stresses_linear_average(rvals)

def mechanical_stresses_cyl_average(r):
    sigma_r = A - 2 * B /(Ro**2 - Ri**2) * np.log(Ro/Ri) * np.ones_like(r)
    sigma_t = A + 2 * B /(Ro**2 - Ri**2) * np.log(Ro/Ri) * np.ones_like(r)
    sigma_z = A * np.ones_like(r) # GPS
    return sigma_r, sigma_t, sigma_z

sigma_r_mech_cyl, sigma_t_mech_cyl, sigma_z_mech_cyl = mechanical_stresses_cyl_average(rvals)

# ----
# Plot
# ----
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(rvals, sigma_r_mech / 1e6, "b-", label=r"$\sigma_r$ (Lamè)")
ax.plot(rvals, sigma_t_mech / 1e6, "g-", label=r"$\sigma_\vartheta$ (Lamè)")
ax.plot(rvals, sigma_z_mech / 1e6, "k-", label=r"$\sigma_z$ (Lamè)")

ax.plot(rvals, sigma_r_mech_lin / 1e6, "b:", label=r"$\langle \sigma_r \rangle$_lin")
ax.plot(rvals, sigma_t_mech_lin / 1e6, "g:", label=r"$\langle \sigma_\vartheta \rangle$_lin")
ax.plot(rvals, sigma_z_mech_lin / 1e6, "k:", label=r"$\langle \sigma_\vartheta \rangle$_lin")

ax.plot(rvals, sigma_r_mech_cyl / 1e6, "b.", label=r"$\langle \sigma_r \rangle$_cyl")
ax.plot(rvals, sigma_t_mech_cyl / 1e6, "g.", label=r"$\langle \sigma_r \rangle$_cyl")
ax.plot(rvals, sigma_z_mech_cyl / 1e6, "k.", label=r"$\langle \sigma_r \rangle$_cyl")

ax.set_title(f"Mechanical loading: cylindrical shell. D/t = {2*Ri/t:.2f}")
ax.set_xlabel("r (m)")
ax.set_ylabel("Stress (MPa)")
ax.grid(True)
ax.legend()

plt.show()
```

## Example

### Internal pressure contribution ($p_i = p, p_o = 0$)
Using the formulas from the Lamé model, the tangential stress due solely to $p_i$ is:
$$\sigma_{\theta, i} = A_i + \frac{B_i}{r^2} = p \frac{R_i^2}{R_o^2 - R_i^2} \left( 1 + \frac{R_o^2}{r^2} \right)$$
This value is always **positive** (tensile stress).

### External pressure contribution ($p_i = 0, p_o = p$)
The stress due solely to $p_o$ is:
$$\sigma_{\theta, o} = A_o + \frac{B_o}{r^2} = -p \frac{R_o^2}{R_o^2 - R_i^2} \left( 1 + \frac{R_i^2}{r^2} \right)$$
This value is always **negative** (compressive stress).

### The sum / the hydrostatic case
If you sum the two contributions, thanks to the linearity of the system:
$$\sigma_{\theta, tot} = \sigma_{\theta, i} + \sigma_{\theta, o}$$

By carrying out the algebra, all the terms depending on the radius $1/r^2$ cancel out perfectly, and the constant result is:
$$\sigma_{\theta, tot} = \frac{p R_i^2 - p R_o^2 + \frac{p R_i^2 R_o^2}{r^2} - \frac{p R_i^2 R_o^2}{r^2}}{R_o^2 - R_i^2} = \frac{p(R_i^2 - R_o^2)}{R_o^2 - R_i^2} = -p$$

This proves that even though the individual contributions vary across the thickness, their superposition in a hydrostatic state results in a perfectly **uniform stress field**.


## Mariotte solutions:
If we solve the force balance on a cylinder, pressurized from inside and outside, we get for the hoop force balance:
$$
2 \sigma_\vartheta (R_o - R_i) \text{dz} = \int_0^\pi p_i R_i \sin(\theta) \text{d} \theta \text{dz} - \int_0^\pi p_o R_o \sin(\theta) \text{d} \theta \text{dz}
$$

$$
2 \sigma_\vartheta (R_o - R_i) = 2 p_i R_i - 2 p_o R_o
$$

that gives:

$$
\sigma_\vartheta = \dfrac{p_i R_i}{t} - \dfrac{p_o Ro}{t}
$$

Then, for the axial force balance:

$$
\sigma_z \pi (R_o^2 - R_i^2) = p_i \pi R_i^2 - p_o \pi R_o^2
$$

that gives:

$$
\sigma_z = \dfrac{ p_i R_i^2 - p_o R_o^2}{R_o^2 - R_i^2}
$$

Lastly, the radial stress is:

$$
\sigma_r = \frac{- p_i R_i - p_o R_o}{R_o + R_i}
$$

If $r_i/t > 5 \div 7.5$, the cylinder can be considered thin-walled, and the following approximation is valid:

$$
R = r_i = r_e = r_m
$$

In this case, the relations simplify

$$
\sigma_r = - \frac{p_i + p_o}{2}
$$

$$
\sigma_\vartheta = p_i \frac{R}{t} - p_o \frac{R}{t}
$$

$$
\sigma_z = p_i \frac{R}{2t} - p_o \frac{R}{2t}
$$

When $p_e \neq 0$, the risk of collapse by instability must be considered.


```python
def mariotte_stresses(r):
    R=(Ri+Ro)/2
    sigma_r = -(pi+po)/2 * np.ones_like(r)
    sigma_t = (pi*R-po*R)/t * np.ones_like(r)
    sigma_z = (pi*R-po*R)/(2*t) * np.ones_like(r)
    return sigma_r, sigma_t, sigma_z

sigma_r_mech_M, sigma_t_mech_M, sigma_z_mech_M = mariotte_stresses(rvals)

# ----
# Plot
# ----
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(rvals, sigma_r_mech / 1e6, "b-", label=r"$\sigma_r$ (Lamè)")
ax.plot(rvals, sigma_t_mech / 1e6, "g-", label=r"$\sigma_\vartheta$ (Lamè)")
ax.plot(rvals, sigma_z_mech / 1e6, "k-", label=r"$\sigma_z$ (Lamè)")

ax.plot(rvals, sigma_r_mech_M / 1e6, "b:", label=r"$\sigma_r$ (Mariotte)")
ax.plot(rvals, sigma_t_mech_M / 1e6, "g:", label=r"$\sigma_\vartheta$ (Mariotte)")
ax.plot(rvals, sigma_z_mech_M / 1e6, "k:", label=r"$\sigma_z$ (Mariotte)")

ax.set_title(f"Mechanical loading: cylindrical shell. D/t = {2*Ri/t:.2f}")
ax.set_xlabel("r (m)")
ax.set_ylabel("Stress (MPa)")
ax.grid(True)
ax.legend()

plt.show()
```

Dimensional variations $\Delta r_i, \Delta r_e, \Delta l$ are determined by calculating, through the linear elastic relationship between stress and strain (Hooke’s law), the strains $\varepsilon_t(r_i), \varepsilon_t(r_e), \varepsilon_z$ produced by the stress state defined by $\sigma_r, \sigma_\vartheta, \sigma_z$:

$$
\Delta r = u = \varepsilon_t(r)  r, \quad
\Delta r_i = \varepsilon_t(r_i)  r_i, \quad
\Delta r_e = \varepsilon_t(r_e)  r_e
$$

Dimensional variations for a cylinder of length $z$ closed at the ends are:

$$
\Delta R = \dfrac{p_i - p_o}{Et}R^2 \left( 1 - \dfrac{\nu}{2} \right)
$$

$$
\Delta z = \dfrac{p_i - p_o}{Es}Rz \left( \dfrac{1}{2} - \nu \right)
$$

### Comparison between linear and cylindrical averages

The discrepancy between the two averaging methods depends solely on the geometric ratio $R_o/R_i$. The ratio is:

$$
\mathcal{f} = \dfrac{2 R_i R_o \ln(R_o/R_i)}{R_o^2 - R_i^2}
$$

As the cylinder becomes thinner ($R_o \to R_i$), $\mathcal{f} \to 1$, confirming that Mariotte's linear approximation becomes equivalent to the volumetric average.

```python
# slenderness ratio = Di/t
SR = np.linspace(1e-1, 50, 100)

# geometric_ratio = 2*Ri_array*Ro_array*np.log(Ro_array/Ri_array) / (Ro_array**2 - Ri_array**2)
geometric_ratio = SR * (SR/2 + 1) / (SR + 1) * np.log(1+2/SR)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(SR, geometric_ratio, color='orange', label='Cylindrical-over-linear average')

ax.set_xlabel('Slenderness ratio (/)')
ax.set_ylabel('Ratio (/)')
ax.grid(True)
ax.legend()

plt.show()
```

### Thermal stresses in a cylindrical shell

The analytical thermal stresses in a cylindrical shell that develop after a radial temperature profile $T(r)$ are expressed by:

$$
\sigma_r(r) =
\frac{\alpha E}{1-\nu}\frac{1}{r^2}
\left[
\frac{r^2 - R_i^2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r')\, r' \,dr'
-
\int_{R_i}^{r} T(r')\, r' \,dr'
\right]
$$

$$
\sigma_\theta(r) =
\frac{\alpha E}{1-\nu}
\left[
\frac{1 + (R_i/r)^2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r')\, r' \,dr'
+
\frac{1}{r^2} \int_{R_i}^{r} T(r')\, r' \,dr'
- T(r)
\right]
$$

$$
\sigma_z(r) =
\frac{\alpha E}{1-\nu}
\left[
\frac{2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r')\, r' \,dr'
- T(r)
\right]
$$

The formulas above are valid for an infinitely long tube ($\varepsilon_{zz} = 0$). Otherwise, something else is needed (e.g., GPS, $F_z = \int_{r_i}^{r_e} \sigma_z\ 2\pi r\ dr$)



```python
from scipy.integrate import quad
from scipy.optimize import fsolve

alpha = 1.2e-5 # linear thermal expansion coefficient

Ti_val = 100.0 
To_val = 100.0
Tmin = 20.0

def T_profile(r):
    r_mid = (Ri + Ro) / 2
    k = (Ti_val - Tmin) / (Ri - r_mid)**2

    # Temperature profile: sad (concave) parabola
    # The central ring is the hottest one. It expands towards the outside.
    # Therefore, the outer rings are compressed by it, while the inner rings are in tensile state.
    return Tmin - k * (r - r_mid)**2

    # Temperature profile: happy (convex) parabola
    # The central ring is the coldest one.
    # The inner (hot) rings expand towards the outside, but they are compressed due to the cold central ring.
    # The outer (hot) rings expand towards the outside, they are in tensile state.
    # return Tmin + k * (r - r_mid)**2

integral_Tr, _ = quad(lambda r: T_profile(r) * r, Ri, Ro)
T_mean = (2.0 / (Ro**2 - Ri**2)) * integral_Tr
r1 = fsolve(lambda r: T_profile(r) - T_mean, Ri + 50)[0] * 1e3
r2 = fsolve(lambda r: T_profile(r) - T_mean, Ro - 50)[0] * 1e3
r3 = fsolve(lambda r: T_profile(r) - np.max(T_profile(r)), (Ri+Ro)/2)[0] * 1e3

# --- Thermal stresses ---
def stresses_thermal_integral(r):
    I_re, _ = quad(lambda rho: T_profile(rho) * rho, Ri, Ro)
    I_r, _ = quad(lambda rho: T_profile(rho) * rho, Ri, r)
    
    Tr = T_profile(r)
    factor = (alpha * E) / (1 - nu)
    
    # Lamé-Gadolin
    sigma_r = (factor / r**2) * (((r**2 - Ri**2) / (Ro**2 - Ri**2)) * I_re - I_r)
    sigma_t = (factor / r**2) * (((r**2 + Ri**2) / (Ro**2 - Ri**2)) * I_re + I_r - Tr * r**2)    
    sigma_z = factor * ((2 / (Ro**2 - Ri**2)) * I_re - Tr) # plane strain
    
    return sigma_r, sigma_t, sigma_z

results = [stresses_thermal_integral(r) for r in rvals]
sigma_r_th_int, sigma_t_th_int, sigma_z_th_int = zip(*results)

# --- Plotting ---
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(rvals * 1e3, np.array(sigma_r_th_int) / 1e6, "b-", lw=2.5, label=r"$\sigma_r$")
ax1.plot(rvals * 1e3, np.array(sigma_t_th_int) / 1e6, "g-", lw=2, label=r"$\sigma_\vartheta$")
ax1.plot(rvals * 1e3, np.array(sigma_z_th_int) / 1e6, "k-", lw=2, label=r"$\sigma_z$")

ax1.vlines([r1, r2, r3], -50, 50, colors='gray', linestyles='dashdot', alpha=0.5)
ax1.axhline(0, color='black', linewidth=1, alpha=0.5)
ax1.set_xlabel("r (mm)")
ax1.set_ylabel("Stress (MPa)")
ax1.grid(True, linestyle=':', alpha=0.6)

ax2 = ax1.twinx()
ax2.plot(rvals * 1e3, [T_profile(r) for r in rvals], "r--", label="Temperature")
ax2.set_ylabel("Temperature (°C)", color="r")
ax2.tick_params(axis='y', labelcolor="r")

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="best")

plt.tight_layout()
plt.show()
```

### Thermal stresses and membrane approximation

For thin cylinderical shells, the thermal hoop stress is approximated by:

$$
\sigma_\theta^{\text{thin}}(r) =
\frac{\alpha E}{1-\nu} \left( \bar{T} - T(r) \right)
$$

where the average temperature is

$$
\bar{T} = \frac{2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r)\,r\,dr .
$$

In the thin-wall limit:

$$
\sigma_\theta^{\text{thin}}(r) = \sigma_z(r)
$$

because both axial and hoop stresses respond only to the average mismatch between $\bar{T}$ and $T(r)$.
