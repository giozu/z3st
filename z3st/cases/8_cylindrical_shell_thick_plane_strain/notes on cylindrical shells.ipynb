# Pressure vessels

## Cylindrical vessels and thick-walled tubes

### Internal and external pressure

The mechanical stresses according to **Lamé** for an infinitely long cylinder closed at the ends are

$$
\sigma_r = -p_i \frac{(r_e/r)^2 - 1}{a^2 - 1} - p_e \frac{a^2 - (r_e/r)^2}{a^2 - 1}
$$

$$
\sigma_t = p_i \frac{(r_e/r)^2 + 1}{a^2 - 1} - p_e \frac{a^2 + (r_e/r)^2}{a^2 - 1}
$$

$$
\sigma_z = \frac{\sigma_r + \sigma_t}{2} = p_i \frac{1}{a^2 - 1} - p_e \frac{a^2}{a^2 - 1}
$$

where $a = r_e / r_i$.
If the longitudinal pressure is zero or balanced by external action, then $\sigma_z = 0$.

In the case of a single acting pressure (internal or external), the stresses reach their maximum and minimum values at the boundaries.
In the general case:

* for $r = r_i$:

$$
\sigma_r = -p_i, \quad
\sigma_t = p_i \frac{a^2 + 1}{a^2 - 1} - p_e \frac{2 a^2}{a^2 - 1}, \quad
\sigma_z = p_i \frac{1}{a^2 - 1} - p_e \frac{a^2}{a^2 - 1}
$$

* for $r = r_e$:

$$
\sigma_r = -p_e, \quad
\sigma_t = p_i \frac{2}{a^2 - 1} - p_e \frac{a^2 + 1}{a^2 - 1}, \quad
\sigma_z = p_i \frac{1}{a^2 - 1} - p_e \frac{a^2}{a^2 - 1}
$$

Dimensional variations $\Delta r_i, \Delta r_e, \Delta l$ are determined by calculating, through the linear elastic relationship between stress and strain (Hooke’s law), the strains $\varepsilon_t(r_i), \varepsilon_t(r_e), \varepsilon_z$ produced by the stress state defined by $\sigma_r, \sigma_t, \sigma_z$:

$$
\Delta r = u = \varepsilon_t(r)  r, \quad
\Delta r_i = \varepsilon_t(r_i)  r_i, \quad
\Delta r_e = \varepsilon_t(r_e)  r_e
$$

---

### Thermal loading

Neglecting thermal accelerations, the elastic stresses for an infinitely long cylinder are:

$$
\sigma_r = \frac{\alpha E}{1 - \nu} \frac{1}{r^2}
\left( \frac{(r/r_i)^2 - 1}{a^2 - 1} \int_{r_i}^{r_e} T r dr -
\int_{r_i}^r T r dr \right)
$$

$$
\sigma_t = \frac{\alpha E}{1 - \nu} \frac{1}{r^2}
\left( \frac{(r/r_i)^2+1}{a^2-1} \int_{r_i}^{r_e} T r dr + \int_{r_i}^r T r dr  -T r^2 \right)
$$

$$
\sigma_z = \frac{\alpha E}{1 - \nu}
\left( \frac{2/r_i^2}{a^2-1} \int_{r_i}^{r_e} T r dr -T \right)
$$

In the case of a steady-state temperature distribution, with boundary conditions $T = T_i$ at $r = r_i$ and $T = T_e$ at $r = r_e$:

$$
T(r) = T_e + \frac{T_i - T_e}{\ln a} \ln \frac{r_e}{r}
$$

the stresses become:

$$
\sigma_r = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu) \ln a}
\left[ -\ln \frac{r_e}{r} - \frac{1}{a^2 - 1} \left(1 - \frac{r_e^2}{r^2}\right) \ln a \right]
$$

$$
\sigma_t = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu) \ln a}
\left[ 1 - \ln \frac{r_e}{r} - \frac{1}{a^2 - 1} \left(1 + \frac{r_e^2}{r^2}\right) \ln a \right]
$$

$$
\sigma_z = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu) \ln a}
\left[ 1 - 2 \ln \frac{r_e}{r} - \frac{2}{a^2 - 1} \ln a \right]
$$

Note that $\sigma_r = 0$ for $r = r_i$ and $r = r_e$, while for any other $r$, with $T_i > T_e$, one has $\sigma_r < 0$.

* for $r = r_i$:

$$
\sigma_t = \sigma_z = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu) \ln a}
\left[ 1 - \frac{2 a^2}{a^2 - 1} \ln a \right] < 0
$$

* for $r = r_e$:

$$
\sigma_t = \sigma_z = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu) \ln a}
\left[ 1 - \frac{2}{a^2 - 1} \ln a \right] > 0
$$

Dimensional variations $\Delta r, \Delta r_e, \Delta l$ are obtained by calculating, through Hooke’s law, the strains $\varepsilon_t(r_i), \varepsilon_t(r_e)$ due to the known stress state $\sigma_r, \sigma_t, \sigma_z$:

$$
\Delta r = u = \varepsilon_t(r) , r, \quad
\Delta r_i = \varepsilon_t(r_i) , r_i, \quad
\Delta r_e = \varepsilon_t(r_e) , r_e
$$

The formulas above are valid for an infinitely long tube. If this condition is not satisfied, the problem must be treated using the general procedures of elasticity theory.

---

## Cylindrical vessels and thin-walled tubes

If $r_i/t > 5 \div 7.5$, the cylinder can be considered thin-walled, and the following approximation is valid:

$$
R = r_i = r_e = r_m
$$

In this case, the relations simplify considerably.

### Under internal and external pressure

One can use the **Mariotte** formulas:

* for $r = r_i$: $\sigma_r = -p_i$
* for $r = r_e$: $\sigma_r = -p_e$

$$
\sigma_t = p_i \frac{R}{t} - p_e \frac{R}{t}
$$

$$
\sigma_r = - \frac{p_i+p_e}{2}
$$

If an axial stress exists due solely to pressure, it is:

$$
\sigma_z = p_i \frac{R}{2t} - p_e \frac{R}{2t}
$$

When $p_e \neq 0$, the risk of collapse by instability must be considered.
Dimensional variations for a cylinder of length $z$ closed at the ends are:

$$
\Delta R = \dfrac{p_i - p_e}{Es}R^2 \left( 1 - \dfrac{\nu}{2} \right)
$$

$$
\Delta z = \dfrac{p_i - p_e}{Es}Rz \left( \dfrac{1}{2} - \nu \right)
$$

---

### Thermal loading

Expanding in a power series:

$$
\ln \frac{r_e}{r_i} = \ln \left(1 + \frac{t}{r_i}\right)
$$

and neglecting the terms $(t/r_i)^n$ with $n \geq 2$, one obtains:

* for $r = r_i$:

$$
\sigma_t = \sigma_z = -\frac{\alpha E (T_i - T_e)}{2 (1 - \nu)}
\left(1 + \frac{1}{3}\frac{t}{r_i}\right)
$$

* for $r = r_e$:

$$
\sigma_t = \sigma_z = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu)}
\left(1 - \frac{1}{3}\frac{t}{r_i}\right)
$$

If $t/r_i$ is neglected compared to unity, then:

$$
T = T_e + (T_i - T_e) \frac{r_e - r}{t}
$$

which expresses a linear variation of temperature across the tube wall.

* for $r = r_i$:

$$
\sigma_t = \sigma_z = -\frac{\alpha E (T_i - T_e)}{2 (1 - \nu)}
$$

* for $r = r_e$:

$$
\sigma_t = \sigma_z = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu)}
$$

Finally:

$$
\sigma_t = \sigma_z = \frac{\alpha E (T_i - T_e)}{2 (1 - \nu)}
\left(1 - 2 \frac{r_e - r}{t}\right)
$$

which varies linearly through the wall thickness.

The above formulas, valid for an infinitely long tube, can also be applied to one of finite length by applying at its ends a uniformly distributed moment along the circumference. This moment produces axial stresses equal and opposite to those previously determined due to thermal loading.
Such a moment per unit length is:

$$
M = \frac{1}{12} \frac{\alpha E |T_i - T_e|}{1 - \nu} t^2
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# ------------------------------------------------------------
# Parameters
# ------------------------------------------------------------
E = 210e9        # Young's modulus [Pa]
nu = 0.3         # Poisson's ratio
alpha = 1e-5     # thermal expansion coefficient [1/K]

ri = 0.050       # inner radius (m)
re = 0.065       # outer radius (m)
t  = re - ri
R  = (ri + re)/2
a = re/ri

pi = 10e6        # internal pressure [Pa]
pe = 0           # external pressure [Pa]

Ti = 400.0       # inner temperature [°C]
Te = 100.0       # outer temperature [°C]

rvals = np.linspace(ri, re, 150)

# ------------------------------------------------------------
# Temperature profile (steady conduction)
# ------------------------------------------------------------
def T_profile(r):
    return Te + (Ti - Te)/np.log(a) * np.log(re/r)

# ------------------------------------------------------------
# 1. Thick-walled mechanical (Lamé formulas)
# ------------------------------------------------------------
def stresses_mechanical(r):
    sigma_r = -pi*((re/r)**2 - 1)/(a**2 - 1) - pe*(a**2 - (re/r)**2)/(a**2 - 1)
    sigma_t =  pi*((re/r)**2 + 1)/(a**2 - 1) - pe*(a**2 + (re/r)**2)/(a**2 - 1)
    sigma_z =  pi/(a**2 - 1) - pe*a**2/(a**2 - 1)
    return sigma_r, sigma_t, sigma_z

sigma_r_mech, sigma_t_mech, sigma_z_mech = zip(*[stresses_mechanical(r) for r in rvals])

# ------------------------------------------------------------
# 2. Thick-walled thermal stresses (integral form)
# ------------------------------------------------------------
def stresses_thermal_integral(r):
    I_re, _ = quad(lambda rho: T_profile(rho)*rho, ri, re)
    I_r,  _ = quad(lambda rho: T_profile(rho)*rho, ri, r)
    Tr = T_profile(r)
    sigma_r = (alpha*E/(1-nu))*(1/r**2)*(((r/ri)**2 - 1)/(a**2-1)*I_re - I_r)
    sigma_t = (alpha*E/(1-nu))*(1/r**2)*(((r/ri)**2 + 1)/(a**2-1)*I_re + I_r - Tr*r**2)
    sigma_z = (alpha*E/(1-nu))*((2/ri**2)/(a**2-1)*I_re - Tr)
    return sigma_r, sigma_t, sigma_z

sigma_r_th_int, sigma_t_th_int, sigma_z_th_int = zip(*[stresses_thermal_integral(r) for r in rvals])

# ------------------------------------------------------------
# 3. Thin-walled mechanical
# ------------------------------------------------------------
sigma_t_thin_mech = (pi - pe)*R/t
sigma_r_thin_mech = -(pi + pe)/2
sigma_z_thin_mech = (pi - pe)*R/(2*t)

# ------------------------------------------------------------
# 4. Thin-walled thermal (linear distribution)
# ------------------------------------------------------------
sigma_t_thin_th = (alpha*E*(Ti-Te))/(2*(1-nu))*(1 - 2*(re - rvals)/t)
sigma_z_thin_th = sigma_t_thin_th.copy()

# ------------------------------------------------------------
# Plots (2 subplots: left=mechanical, right=thermal)
# ------------------------------------------------------------
fig, axs = plt.subplots(1, 2, figsize=(12,5))

# --- mechanical comparison ---
axs[0].plot(rvals*1e3, np.array(sigma_r_mech)/1e6, 'b-', label=r'Thick $\sigma_r$')
axs[0].plot(rvals*1e3, np.array(sigma_t_mech)/1e6, 'g-', label=r'Thick $\sigma_t$')
axs[0].axhline(sigma_z_mech[0]/1e6, color='k', label=r'Thick $\sigma_z$')

axs[0].axhline(sigma_r_thin_mech/1e6, color='b', ls=':', label=r'Thin $\sigma_r$')
axs[0].axhline(sigma_t_thin_mech/1e6, color='g', ls=':', label=r'Thin $\sigma_t$')
axs[0].axhline(sigma_z_thin_mech/1e6, color='k', ls=':', label=r'Thin $\sigma_z$')

axs[0].set_title("Mechanical loading: thick vs thin wall")
axs[0].set_xlabel("r (mm)"); axs[0].set_ylabel("Stress (MPa)")
axs[0].grid(True); axs[0].legend()

# --- thermal comparison ---
axs[1].plot(rvals*1e3, np.array(sigma_r_th_int)/1e6, 'b-', label=r'Thick $\sigma_r$')
axs[1].plot(rvals*1e3, np.array(sigma_t_th_int)/1e6, 'g-', label=r'Thick $\sigma_t$')
axs[1].plot(rvals*1e3, np.array(sigma_z_th_int)/1e6, 'k-', label=r'Thick $\sigma_z$')

axs[1].plot(rvals*1e3, sigma_t_thin_th/1e6, color='g', ls=':', label=r'Thin $\sigma_t=\sigma_z$')

ax2 = axs[1].twinx()
line_T, = ax2.plot(rvals*1e3, [T_profile(r) for r in rvals], color='r', label="T")
ax2.set_ylabel("T (°C)", color='k')

lines1, labels1 = axs[1].get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
axs[1].legend(lines1 + [line_T], labels1 + ["T"], loc="upper right")

axs[1].set_title("Thermal loading: thick vs thin wall")
axs[1].set_xlabel("r (mm)"); axs[1].set_ylabel("Stress (MPa)")
axs[1].grid(True)

slenderness = ri/t
label_text = f"$r_i/t$ = {slenderness:.2f}\n{'thin tube' if slenderness>5 else 'thick wall'}"

for ax in axs:
    ax.text(0.05, 0.70, label_text,
            transform=ax.transAxes,
            fontsize=10, va="top", ha="left",
            bbox=dict(boxstyle="round", facecolor="w", alpha=1.0))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.integrate import quad

def err(thin, thick):
    return np.abs(thin - thick)

def radial_avg(func):
    num, _ = quad(lambda r: func(r)*r, ri, re)
    return 2/(re**2 - ri**2)*num

rows = []

# --- mechanical (radial average) ---
sr_thick_avg = radial_avg(lambda r: stresses_mechanical(r)[0]/1e6)
st_thick_avg = radial_avg(lambda r: stresses_mechanical(r)[1]/1e6)
sz_thick_avg = radial_avg(lambda r: stresses_mechanical(r)[2]/1e6)

rows += [
    {"case":"mechanical","component":"sigma_r",
     "thin (MPa)":sigma_r_thin_mech/1e6,"thick avg (MPa)":sr_thick_avg,
     "abs. err. (MPa)":err(sigma_r_thin_mech/1e6, sr_thick_avg)},
    {"case":"mechanical","component":"sigma_t",
     "thin (MPa)":sigma_t_thin_mech/1e6,"thick avg (MPa)":st_thick_avg,
     "abs. err. (MPa)":err(sigma_t_thin_mech/1e6, st_thick_avg)},
    {"case":"mechanical","component":"sigma_z",
     "thin (MPa)":sigma_z_thin_mech/1e6,"thick avg (MPa)":sz_thick_avg,
     "abs. err. (MPa)":err(sigma_z_thin_mech/1e6, sz_thick_avg)},
]

# --- thermal (radial average) ---
sr_thick_avg = radial_avg(lambda r: stresses_thermal_integral(r)[0]/1e6)
st_thick_avg = radial_avg(lambda r: stresses_thermal_integral(r)[1]/1e6)
sz_thick_avg = radial_avg(lambda r: stresses_thermal_integral(r)[2]/1e6)

# thin wall thermal: at mid-radius
def thin_thermal(rp):
    st = (alpha*E*(Ti-Te))/(2*(1-nu))*(1 - 2*(re - rp)/t)/1e6
    return 0.0, st, st

sr_thin, st_thin, sz_thin = thin_thermal(R)

rows += [
    {"case":"thermal","component":"sigma_r",
     "thin (MPa)":sr_thin,"thick avg (MPa)":sr_thick_avg,
     "abs. err. (MPa)":err(sr_thin, sr_thick_avg)},
    {"case":"thermal","component":"sigma_t",
     "thin (MPa)":st_thin,"thick avg (MPa)":st_thick_avg,
     "abs. err. (MPa)":err(st_thin, st_thick_avg)},
    {"case":"thermal","component":"sigma_z",
     "thin (MPa)":sz_thin,"thick avg (MPa)":sz_thick_avg,
     "abs. err. (MPa)":err(sz_thin, sz_thick_avg)},
]

# build dataframe
df = pd.DataFrame(rows, columns=["case","component","thin (MPa)","thick avg (MPa)","abs. err. (MPa)"])
pd.set_option('display.float_format', lambda x: f"{x:,.3f}")

print("\nAbsolute error of thin-wall approximation vs avg thick-wall stresses:")
print(df.to_string(index=False))
```

### Thermal stress analysis in a hollow cylinder (no pressure)

This script computes the analytical thermal stresses in a thick-walled cylinder subjected to a radial temperature gradient $T(r)$, using the integral formulation:

$$
\sigma_r(r) =
\frac{\alpha E}{1-\nu}\frac{1}{r^2}
\left[
\frac{r^2 - R_i^2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r')\, r' \,dr'
-
\int_{R_i}^{r} T(r')\, r' \,dr'
\right]
$$

$$
\sigma_\theta(r) =
\frac{\alpha E}{1-\nu}
\left[
\frac{1 + (R_i/r)^2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r')\, r' \,dr'
+
\frac{1}{r^2} \int_{R_i}^{r} T(r')\, r' \,dr'
- T(r)
\right]
$$

$$
\sigma_z(r) =
\frac{\alpha E}{1-\nu}
\left[
\frac{2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r')\, r' \,dr'
- T(r)
\right]
$$

---

### Thin-wall (membrane) approximation

For slender cylinders $\varepsilon = (R_o - R_i)/R_i \ll 1$, the hoop stress is approximated by:

$$
\sigma_\theta^{\text{thin}}(r) =
\frac{\alpha E}{1-\nu} \left( \bar{T} - T(r) \right)
$$

where the area-weighted mean temperature is

$$
\bar{T} = \frac{2}{R_o^2 - R_i^2} \int_{R_i}^{R_o} T(r)\,r\,dr .
$$

In the thin-wall limit:

$$
\sigma_\theta^{\text{thin}}(r) = \sigma_z(r)
$$

because both axial and hoop stresses respond only to the average mismatch between $\bar{T}$ and $T(r)$.

---

### Meaning of integral terms

We define:

| Symbol | Definition | Meaning |
|--------|------------|---------|
| $I_{\text{global}}$ | $\displaystyle \int_{R_i}^{R_o} T(r')\,r'\,dr'$ | Global (average) thermal loading |
| $I_{\text{local}}(r)$ | $\displaystyle \int_{R_i}^{r} T(r')\,r'\,dr'$ | Local accumulation up to radius $r$ |
| $P(r)$ | $T(r)\,r^2$ | Pointwise term in $\sigma_\theta$ |

In the code we compute the ratios using the global contribution actually appearing in $\sigma_\theta$:

$$
G_\theta(r) = \frac{1 + (R_i/r)^2}{R_o^2 - R_i^2}\, I_{\text{global}}
$$

Then:

$$
\frac{L}{G_\theta} = \frac{I_{\text{local}}(r)}{G_\theta(r)},
\qquad
\frac{T(r)\,r^2}{G_\theta(r)}
$$

These ratios tell us:

- If $L/G_\theta \ll 1$ → local effects are negligible → membrane solution is valid  
- If $(T r^2)/G_\theta \approx 1$ → global and point terms balance → stresses follow the thin-wall law

---

### Slenderness ratio

$$
\text{slenderness} = \frac{R_i}{R_o - R_i}
$$

This geometric ratio tells us if the cylinder behaves like thick or thin:

| Slenderness | Interpretation |
|-------------|----------------|
| $< 5$       | Thick cylinder |
| $5$–$10$    | Transitional   |
| $> 10$      | Thin-wall      |

---

### Output

The script provides:

- Stresses $\sigma_r(r)$, $\sigma_\theta(r)$, $\sigma_z(r)$  
- Temperature distribution $T(r)$  
- Ratios $L/G_\theta$ and $(T r^2)/G_\theta$  
- Slenderness $R_i/(R_o-R_i)$ and RMSE vs thin-wall stress

Plots include:
- **Figure 1:** Thermal stresses (MPa)  
- **Figure 2:** Temperature profile  
- **Figure 3:** Ratio analysis (L/G and $(Tr^2)/G$)


```python
#!/usr/bin/env python3
"""
Thermal stress in a hollow cylinder (no pressure)
Using the analytical_thermal_stress formulation.

Includes:
- Stress computation (sigma_r, sigma_t, sigma_z)
- Thin-wall approximation for sigma_theta
- Integral term ratios (L/G and T*r^2/G)
- Plots for stresses (last case), ratios, temperature
- Parametric study: RMSE vs slenderness (log–log)
"""

import numpy as np
import matplotlib.pyplot as plt

# Geometry & material
Ri, Ro = 2.0, 2.001     # m
E = 1.77e11             # Pa
nu = 0.3
alpha = 1.7e-5          # 1/K
Ti, To = 490.0, 480.0   # K

n_points = 10_000

def analytic_T(r):
    return Ti + (To - Ti) * np.log(r / Ri) / np.log(Ro / Ri)

def T_bar_from_Iglobal(I_global):
    return 2.0 * I_global / (Ro**2 - Ri**2)

def sigma_theta_thin(T_r, I_global):
    Tbar = T_bar_from_Iglobal(I_global)
    pref = alpha * E / (1.0 - nu)
    return pref * (Tbar - T_r), Tbar

def analytical_thermal_stress(r):
    r_star = np.linspace(Ri, Ro, n_points)
    T_star = analytic_T(r_star)
    I_global = np.trapezoid(T_star * r_star, r_star)

    sigma_r = np.zeros_like(r)
    sigma_t = np.zeros_like(r)
    sigma_t_approx = np.zeros_like(r)
    sigma_z = np.zeros_like(r)

    for i, ri in enumerate(r):
        mask = r_star <= ri
        I_local = np.trapezoid(T_star[mask] * r_star[mask], r_star[mask])

        sigma_r[i] = (alpha * E / (1 - nu)) * ( ((ri**2 - Ri**2)/(Ro**2 - Ri**2)) * I_global - I_local ) / (ri**2)
        sigma_t[i] = (alpha * E / (1 - nu)) * ( ((1 + (Ri/ri)**2)/(Ro**2 - Ri**2)) * I_global + I_local / (ri**2) - analytic_T(ri) )
        sigma_z[i] = (alpha * E / (1 - nu)) * ( (2/(Ro**2 - Ri**2)) * I_global - analytic_T(ri) )

    return sigma_r, sigma_t, sigma_z, I_global

def compute_term_ratios(r, I_global):
    r_star = np.linspace(Ri, Ro, n_points)
    T_star = analytic_T(r_star)
    ratios = []
    for ri in r:
        mask = r_star <= ri
        I_local = np.trapezoid(T_star[mask] * r_star[mask], r_star[mask])
        G_term = ((1 + (Ri/ri)**2)/(Ro**2 - Ri**2)) * I_global
        P_term = analytic_T(ri) * ri**2
        ratios.append((ri, I_local, G_term, P_term, I_local / G_term, P_term / G_term))
    return np.array(ratios)

if __name__ == "__main__":
    results = []

    slenderness_values = np.array([1, 2, 3, 5, 8, 10, 15, 20, 30, 50, 100], dtype=float)
    Ro_initial = Ro

    for slender in slenderness_values:
        t = Ri / slender
        Ro = Ri + t

        r = np.linspace(Ri, Ro, 300)
        sigma_r, sigma_t, sigma_z, I_global = analytical_thermal_stress(r)

        sigma_t_thin, Tbar = sigma_theta_thin(analytic_T(r), I_global)
        rmse = np.sqrt(np.mean((sigma_t - sigma_t_thin)**2))
        rel_rmse = rmse / (np.sqrt(np.mean(sigma_t**2)))

        ratios = compute_term_ratios(r, I_global)
        max_LG = float(np.max(np.abs(ratios[:, 4])))
        max_Tr2G = float(np.max(np.abs(ratios[:, 5])))

        results.append({
            "slenderness": float(slender),
            "t (thickness)": float(t),
            "Ro": float(Ro),
            "Tbar [K]": float(Tbar),
            "RMSE_sigma_theta [Pa]": float(rmse),
            "Rel_RMSE [%]": float(rel_rmse * 100.0),
            "max(L/G)": max_LG,
            "max(Tr^2/G)": max_Tr2G
        })

    print("\n=== Thin-wall approximation validity (σθ) ===")
    print(f"{'Ri/t':>10} | {'t [m]':>8} | {'RMSE [%]':>10} | {'max(L/G)':>10} | {'max(Tr²/G)':>12} | {'T̄ [K]':>8}")
    print("-" * 70)

    for row in results:
        print(f"{row['slenderness']:10.1f} | "
            f"{row['t (thickness)']:8.4f} | "
            f"{row['Rel_RMSE [%]']:10.3f} | "
            f"{row['max(L/G)']:10.4f} | "
            f"{row['max(Tr^2/G)']:12.4f} | "
            f"{row['Tbar [K]']:8.2f}")

    slender_list = np.array([row["slenderness"] for row in results], dtype=float)
    rmse_list = np.array([row["Rel_RMSE [%]"] for row in results], dtype=float)  # %
    LG_list = np.array([row["max(L/G)"] for row in results], dtype=float)

    # avoid log(0)
    rmse_list[rmse_list <= 0] = 1e-16
    LG_list[LG_list <= 0] = 1e-16

    # Case to plot
    slender = 20
    t = Ri / slender
    Ro = Ri + t
    r = np.linspace(Ri, Ro, 300)
    sigma_r, sigma_t, sigma_z, I_global = analytical_thermal_stress(r)
    sigma_t_thin, Tbar = sigma_theta_thin(analytic_T(r), I_global)
    ratios = compute_term_ratios(r, I_global)

    # plot stresses
    fig, ax1 = plt.subplots()
    ax1.plot(r, sigma_t/1e6, label="σθ (exact)")
    ax1.plot(r, sigma_r/1e6, '--', label="σr")
    ax1.plot(r, sigma_z/1e6, '-.', label="σz")
    ax1.plot(r, sigma_t_thin/1e6, ':', label="σθ (thin-wall)")
    ax1.set_xlabel("r (m)")
    ax1.set_ylabel("stress (MPa)")
    ax1.set_title(f"thermal stresses, Sr = {slender}")
    ax1.grid(True)
    ax1.legend()

    # temperature
    fig2, ax2 = plt.subplots()
    ax2.plot(r, analytic_T(r), label="T(r) [K]")
    ax2.set_xlabel("r (m)")
    ax2.set_ylabel("temperature (K)")
    ax2.set_title(f"temperature distribution, Sr = {slender}")
    ax2.grid(True)
    ax2.legend()

    # ratios
    fig3, ax3 = plt.subplots()
    ax3.plot(ratios[:,0], ratios[:,4], label="L / Gθ")
    ax3.plot(ratios[:,0], ratios[:,5], label="(T r²) / Gθ")
    ax3.set_xlabel("r (m)")
    ax3.set_ylabel("ratio (-)")
    ax3.set_title(f"integral term ratios, Sr = {slender}")
    ax3.grid(True)
    ax3.legend()

    # log–log: rmse vs slenderness
    fig4, ax4 = plt.subplots()
    ax4.loglog(slender_list, rmse_list, marker='o', label="Rel. RMSE σθ (%)")
    ax4.set_xlabel("slenderness Ri/t (log)")
    ax4.set_ylabel("relative RMSE σθ (%) (log)")
    ax4.set_title("thin-wall approximation error vs slenderness")
    ax4.grid(True, which="both", ls="--", alpha=0.5)
    ax4.legend()

    # log–log: max(L/G) vs slenderness
    fig5, ax5 = plt.subplots()
    ax5.loglog(slender_list, LG_list, marker='s', label="max(L/Gθ)")
    ax5.set_xlabel("slenderness Ri/t (log)")
    ax5.set_ylabel("max(L/Gθ) (log)")
    ax5.set_title("decay of local contribution vs slenderness")
    ax5.grid(True, which="both", ls="--", alpha=0.5)
    ax5.legend()

    plt.show()
```

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt

# material & thermal setup (modifica se vuoi)
E     = 1.77e11
nu    = 0.30
alpha = 1.7e-5
Ti, To = 490.0, 480.0

Ri = 2.0                          # fisso; variamo m -> Ro = Ri*(1+m)
m_list = np.array([0.30, 0.20, 0.15, 0.10, 0.08, 0.06, 0.05, 0.03, 0.02, 0.015, 0.01, 0.005], dtype=float)

def T_profile(r, Ri, Ro):
    return Ti + (To - Ti) * np.log(r / Ri) / np.log(Ro / Ri)

def exact_stress(r_eval, Ri, Ro):
    # soluzione integrale (nessun carico meccanico)
    r_star = np.linspace(Ri, Ro, 20000)
    T_star = T_profile(r_star, Ri, Ro)
    I_global = np.trapezoid(T_star * r_star, r_star)

    sig_r = np.zeros_like(r_eval)
    sig_t = np.zeros_like(r_eval)
    sig_z = np.zeros_like(r_eval)

    pref = alpha * E / (1.0 - nu)
    for i, r in enumerate(r_eval):
        mask = r_star <= r
        I_local = np.trapezoid(T_star[mask] * r_star[mask], r_star[mask])
        Tr = T_profile(r, Ri, Ro)
        sig_r[i] = pref * ( ((r**2 - Ri**2)/(Ro**2 - Ri**2)) * I_global - I_local ) / r**2
        sig_t[i] = pref * ( ((1 + (Ri/r)**2)/(Ro**2 - Ri**2)) * I_global + I_local / r**2 - Tr )
        sig_z[i] = pref * ( (2/(Ro**2 - Ri**2)) * I_global - Tr )
    return sig_r, sig_t, sig_z, I_global

def thin_sigma_theta(r_eval, Ri, Ro, I_global):
    Tbar = 2.0 * I_global / (Ro**2 - Ri**2)
    pref = alpha * E / (1.0 - nu)
    return pref * (Tbar - T_profile(r_eval, Ri, Ro)), Tbar

def rel_err(a, b):
    denom = np.maximum(np.abs(a), 1e-30)
    return np.abs(a - b) / denom

rows = []
err_vs_m = []
err_vs_S = []

for m in m_list:
    Ro = Ri * (1.0 + m)
    S  = Ri / (Ro - Ri)      # slenderness = Ri/t = 1/m
    r_eval = np.array([Ri, Ro])  # confrontiamo ai bordi

    # esatto
    sr, st, sz, Iglob = exact_stress(r_eval, Ri, Ro)

    # thin-wall membrana
    st_thin, Tbar = thin_sigma_theta(r_eval, Ri, Ro, Iglob)
    sz_thin = st_thin.copy()

    # correzione O(m): fattore (1 ± m/3) ai bordi
    fac_inner = (1.0 + m/3.0)
    fac_outer = (1.0 - m/3.0)
    st_corr = st_thin.copy()
    sz_corr = sz_thin.copy()
    st_corr[0] *= fac_inner
    st_corr[1] *= fac_outer
    sz_corr[0] *= fac_inner
    sz_corr[1] *= fac_outer

    # errori relativi su σθ ai bordi
    e_thin_inner = rel_err(st[0], st_thin[0])
    e_thin_outer = rel_err(st[1], st_thin[1])
    e_corr_inner = rel_err(st[0], st_corr[0])
    e_corr_outer = rel_err(st[1], st_corr[1])

    max_thin = max(e_thin_inner, e_thin_outer)
    max_corr = max(e_corr_inner, e_corr_outer)

    rows.append({
        "m": m, "S=Ri/t": S,
        "err_thin@Ri[%]": 100*e_thin_inner, "err_thin@Ro[%]": 100*e_thin_outer,
        "err_corr@Ri[%]": 100*e_corr_inner, "err_corr@Ro[%]": 100*e_corr_outer,
        "max_err_thin[%]": 100*max_thin, "max_err_corr[%]": 100*max_corr
    })

    err_vs_m.append((m, 100*max_thin, 100*max_corr))
    err_vs_S.append((S, 100*max_thin, 100*max_corr))

# stampa tabella compatta
print("\n=== numerical comparison: exact vs thin vs thin+O(m) (σθ at Ri/Ro) ===")
print(f"{'m':>7} | {'S=Ri/t':>7} | {'thin@Ri%':>9} | {'thin@Ro%':>9} | {'corr@Ri%':>9} | {'corr@Ro%':>9} | {'max thin%':>9} | {'max corr%':>9}")
print("-"*86)
for r in rows:
    print(f"{r['m']:7.4f} | {r['S=Ri/t']:7.2f} | {r['err_thin@Ri[%]']:9.3f} | {r['err_thin@Ro[%]']:9.3f} | "
          f"{r['err_corr@Ri[%]']:9.3f} | {r['err_corr@Ro[%]']:9.3f} | {r['max_err_thin[%]']:9.3f} | {r['max_err_corr[%]']:9.3f}")

# grafici: massimo errore vs m e vs S
err_vs_m = np.array(err_vs_m)
err_vs_S = np.array(err_vs_S)

fig1, ax1 = plt.subplots()
ax1.loglog(err_vs_m[:,0], err_vs_m[:,1], 'o-', label='max err thin (%)')
ax1.loglog(err_vs_m[:,0], err_vs_m[:,2], 's-', label='max err thin+O(m) (%)')
ax1.set_xlabel('m = Ro/Ri - 1 (log)')
ax1.set_ylabel('max relative error σθ (%) (log)')
ax1.set_title('max error vs m')
ax1.grid(True, which='both', ls='--', alpha=0.5)
ax1.legend()

fig2, ax2 = plt.subplots()
ax2.loglog(err_vs_S[:,0], err_vs_S[:,1], 'o-', label='max err thin (%)')
ax2.loglog(err_vs_S[:,0], err_vs_S[:,2], 's-', label='max err thin+O(m) (%)')
ax2.set_xlabel('S = Ri/t (log)')
ax2.set_ylabel('max relative error σθ (%) (log)')
ax2.set_title('max error vs slenderness S')
ax2.grid(True, which='both', ls='--', alpha=0.5)
ax2.legend()

plt.show()
```
